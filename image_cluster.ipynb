```python
import pandas as pd
import numpy as np
import math

from sklearn.cluster import KMeans
from skimage.transform import resize
import tensorflow as tf
from tensorflow import keras

import matplotlib.pyplot as plt
```

```python
total_x = np.load("331x331_total_x.npy")

print(np.shape(total_x))

total_x = total_x.reshape((total_x.shape[0], -1))

print(np.shape(total_x))
```

```python
kmeans = KMeans(n_clusters=10, n_init=10)

y_pred_kmeans = kmeans.fit_predict(total_x)
```

```python
np.save("y_pred_kmeans.npy", y_pred_kmeans)
```

```python
print(y_pred_kmeans[250:400])
```

```python
x_kmeans = []
for i in range(10):
  x_kmeans.append([])
for i in range(len(y_pred_kmeans)):
  x_kmeans[y_pred_kmeans[i]].append(i)

for i in range(10):
  print(x_kmeans[i][0])
```

```python
pre_x = np.load("331x331_total_x.npy")

for i in range(10):
  id = x_kmeans[i][0]
  plt.figure(figsize=(5, 5))
  plt.imshow(pre_x[id])
```

```python
base_model = keras.applications.MobileNetV2(
    weights='imagenet',  # Load weights pre-trained on ImageNet.
    # input_shape=(331, 331, 3),
    include_top=True)

base_model.trainable = False
```

```python
base_model.summary()
```

```python
inputs = keras.Input(shape=(224, 224, 3))
outputs = base_model(inputs)

model = keras.Model(inputs, outputs)
```

```python
model.summary()
```

```python
imgs = []

print(np.shape(pre_x[0]))

for i in range(10):
  id = x_kmeans[i][0]

  imgs.append([resize(pre_x[id], (224, 224, 3))])
  
  # plt.figure(figsize=(5, 5))
  # plt.imshow(imgs[i])
imgs = np.squeeze(imgs)

print(np.shape(imgs))
```

```python
prediction = model.predict(imgs)

print(np.shape(prediction))
print(prediction)
```

```python
for i in range(10):
  max5 = np.argsort(prediction[i])[-5:]
  print(f"Top 5 for picture {i+1} : {max5}")
  print(prediction[i][481])
```
